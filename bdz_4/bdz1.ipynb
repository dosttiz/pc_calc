```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# № 1

x = sp.Symbol('x')
y = sp.Symbol('y')

print('точки пересечения графиков', sp.solve(4 - x**2 - x**2 + 2 * x, x))
```

```python
def f(x):
    return 4 - x**2

def g(x):
    return x**2 - 2 * x

x1 = np.linspace(-1, 2)
plt.plot(x1, f(x1), label='4 - x^2')
plt.plot(x1, g(x1), label='x^2 - 2x')
plt.legend()
plt.grid()
```

```python
# таким образом, f(x) > g(x) на [-1, 2]

S = sp.integrate(4 - x**2 - (x**2 - 2 * x), (x, -1, 2))
print('Площадь:', S)
```

```python
# № 2

phi = np.linspace(0, 2 * np.pi)
r1 = 6 * np.sin(phi)
r2 = 4 * np.sin(phi)
x1 = 6 * np.sin(phi) * np.cos(phi)
y1 = 6 * np.sin(phi) * np.sin(phi)
plt.plot(x1, y1)
x2 = 4 * np.sin(phi) * np.cos(phi)
y2 = 4 * np.sin(phi) * np.sin(phi)
plt.plot(x1, y1, label='r1 = 6 sin(p)')
plt.plot(x2, y2, label='r2 = 4 sin(p)')
plt.legend()
plt.grid()
```

```python
# значит r1 > r2 и S = S_r1 - S_r2
p = sp.Symbol('p')
S = sp.integrate(0.5 * (6 * sp.sin(p))**2, (p, 0, 2 * sp.pi))
S -= sp.integrate(0.5 * (4 * sp.sin(p))**2, (p, 0, 2 * sp.pi))
print('Площадь:', S)
```

```python
# № 3

x = sp.Symbol('x')
x1 = np.linspace(-1, 0)
y1 = np.cosh(x1)
plt.plot(x1, y1)
plt.grid()
L = sp.integrate(sp.sqrt(1 + sp.diff(sp.cosh(x + 1), x)**2), (x, -1, 0))
print('Длина дуги:', float(L))
```

```python
# № 4

t = sp.Symbol('t')
x = sp.cos(t) + sp.log(sp.tan(t / 2))
y = sp.sin(t)
L = sp.integrate(sp.sqrt(sp.diff(x, t)**2 + sp.diff(y, t)**2), (t, sp.pi / 3, sp.pi / 2))
t = np.linspace(np.pi / 3, np.pi / 2)
x1 = np.cos(t) + np.log(np.tan(t / 2))
y1 = np.sin(t)
plt.plot(x1, y1)
plt.grid()
print('Длина дуги:', float(L))
```

```python
# № 5
phi = np.linspace(-np.pi / 4, np.pi / 4)
r = 1 / (1 - np.sin(phi))
x = 1 / (1 - np.sin(phi)) * np.cos(phi)
y = 1 / (1 - np.sin(phi)) * np.sin(phi)
plt.plot(x, y)
plt.grid()
p = sp.Symbol('p')
r = 1 / (1 - sp.sin(p))
L = sp.integrate(sp.sqrt(r**2 + sp.diff(r, p)**2), (p, -sp.pi / 4, sp.pi / 4))
print('Длина дуги', float(L))
```

```python
# № 6

x1 = np.linspace(0, 2)
y1 = x1**2 - 2 * x1 + 1
plt.plot(x1, y1)
plt.axvline(2)
plt.axhline(0)
plt.grid()

# способ 1: V = 2 * pi * \int x * y(x) dx
x = sp.Symbol('x')
y = x**2 - 2 * x + 1
vol = 2 * sp.pi * sp.integrate(x * y, (x, 1, 2))
print('Объем тела вращения:', vol)
```

```python
# способ 2: интегрирование по y, V = V1 - V2

y = sp.Symbol('y')
x = sp.sqrt(y) + 1
v1 = sp.pi * sp.integrate(4, (y, 0, 1))
v2 = sp.pi * sp.integrate(x**2, (y, 0, 1))
print('Объем тела вращения:', v1 - v2)
```
