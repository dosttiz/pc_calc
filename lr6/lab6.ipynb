```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math
```

```python
# upr 1

def plot_solid_fg(f, g, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b)
    plt.plot(x, f(x))
    plt.plot(x, g(x))
    for i in range(n + 1):
        plt.plot(2 * [a + dx * i], [f(a + dx * i), g(a + dx * i)], 'g')
        
def plot_solid_f(f, g, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b)
    plt.plot(x, f(x))
    plt.plot(x, g(x), '--')
    for i in range(n + 1):
        plt.plot(2 * [a + dx * i], [f(a + dx * i), g(a + dx * i)], 'g')

def plot_solid_g(f, g, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b)
    plt.plot(x, f(x), '--')
    plt.plot(x, g(x))
    for i in range(n + 1):
        plt.plot(2 * [a + dx * i], [f(a + dx * i), g(a + dx * i)], 'g')

def plot_no_solid(f, g, a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b)
    plt.plot(x, f(x), '--')
    plt.plot(x, g(x), '--')
    for i in range(n + 1):
        plt.plot(2 * [a + dx * i], [f(a + dx * i), g(a + dx * i)], 'g')
```

```python
# upr 2

plot_solid_fg(lambda x: x**2 - 1, lambda x: x**2 + 1, -2, 2, 10)
plt.grid()
```

```python
# upr 7
# x = exp(|t|) * cos(t), y = sin(t), z = t, t \in [-5pi; 5pi]

t = np.linspace(-5 * np.pi, 5 * np.pi)
x = np.exp(np.abs(t)) * np.cos(t)
y = np.sin(t)
z = t
fig1 = plt.figure()
ax = fig1.add_subplot(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot(x, y, z, '--r')
```
