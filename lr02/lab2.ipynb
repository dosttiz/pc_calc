```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import scipy
import math
```

```python
# upr 1

x = np.linspace(0, np.pi / 2, 1000)
plt.plot(x, x * np.cos(x))
plt.axvline(0)
plt.axvline(np.pi / 2)
plt.axhline(0)
plt.grid()
x = sp.Symbol('x')
print('S =', sp.integrate(x * sp.cos(x), (x, 0, sp.pi / 2)))
```

```python
# upr 2

x1 = np.linspace(-10, 10)
f = x1**2 + 2 * x1
g = 7 - 4 * x1 - x1**2
plt.plot(x1, f)
plt.plot(x1, g, 'r')
plt.grid()
x = sp.Symbol('x')
sol = sp.solve(x**2 + 2 * x - 7 + 4 * x + x**2, x)
s = float(sp.integrate(7 - 4 * x - x**2 - (x**2 + 2 * x), (x, min(sol), max(sol))))
print(sol)
print('s =', s)
```

```python
# 1C

x = np.linspace(-3, 3)
plt.plot(x, x**2 + 1)
plt.plot(x, 3 - x)
plt.axhline(1)
plt.axis('equal')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('1C')

x = sp.Symbol('x')
y1 = x**2 + 1
y2 = 3 - x
s = sp.integrate(y1, (x, 0, 1)) + sp.integrate(y2, (x, 1, 2)) - sp.integrate(1, (x, 0, 2))
print('s =', s)
```

$$ y_1 = x^2 + 1 $$  
$$ y_2 = 3 - x $$  
$$ 
S = \int_0^1 y_1 dx + \int_1^2 y_2 dx - \int_0^2 dx = \\ 
     = \int_1^2 dy \int_{\sqrt{y - 1}}^{3 - y}dx = \\
    = \int_0^1 dx \int_1^{x^2 + 1} dy + \int_1^2 dx \int_1^{3 - x} dy $$

```python
# 2C

t = np.linspace(0, np.pi * 2)
x = np.cos(t)**3
y = np.sin(t)**3
plt.plot(x, y)
plt.grid()
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2C')

t = sp.Symbol('t')
x = sp.cos(t)**3
y = sp.sin(t)**3
s = sp.integrate(y * sp.diff(x, t), (t, sp.pi * 2, 0))
print('s =', s)
```

```python
# 3C

t = np.linspace(0, np.pi * 2)
plt.plot(np.cos(t), np.sin(t))
plt.plot(2 * np.cos(t), 2 * np.sin(t))
plt.plot([0, 2 * np.cos(np.pi / 4)], [0, 2 * np.sin(np.pi / 4)])
plt.plot([0, 2 * np.cos(3 * np.pi / 4)], [0, 2 * np.sin(3 * np.pi / 4)])

plt.axis('equal')
plt.grid()

p = sp.Symbol('p')
s = 0.5 * sp.integrate(4, (p, sp.pi / 4, 3 * sp.pi / 4)) - 0.5 * sp.integrate(1, (p, sp.pi / 4, 3 * sp.pi / 4))
print('s =', s)
```

```python
# 4C

t = sp.Symbol('t')
x = sp.cos(t)**3
y = sp.sin(t)**3
l = sp.integrate(sp.sqrt(sp.diff(x, t)**2 + sp.diff(y, t)**2), (t, 0, sp.pi * 2))
print('L =', float(l))
```
