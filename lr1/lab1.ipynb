```python
# darboux sums

# task3
# a)

def isum_l(f, a, b, n):
    dots = [a + i * (b - a) / n for i in range(n + 1)]
    return [f(k) * (k1 - k) for k, k1 in zip(dots[::2], dots[1::2])]
isum_l(lambda x: x, 1, 2, 4)
```
