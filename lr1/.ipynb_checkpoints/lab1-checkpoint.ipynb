```python
# upr 3
# a)

def isum_left(f, a, b, n):
    dx = (b - a) / n
    div = [a + i * dx for i in range(n)]
    return sum([f(i) * dx for i in div])

isum_left(lambda x: x, 1, 2, 4)
    
```

```python
# upr 3
# b)

def isum_right(f, a, b, n):
    dx = (b - a) / n
    div = [a + i * dx for i in range(n)]
    return sum([f(i + dx) * dx for i in div])

isum_right(lambda x: x, 1, 2, 4)
    
```

```python
# upr 4

def darb_lo(f, a, b, n): 
    dx = (b - a) / n
    div = [a + i * dx for i in range(n)]
    res = []
    for i in div:
        res.append(min(f(np.linspace(i, i + dx, 1000))) * dx)
    return sum(res)

print(r'S(m_i * dx_i) = ', darb_lo(lambda x: x, 1, 2, 4))

def darb_hi(f, a, b, n): 
    dx = (b - a) / n
    div = [a + i * dx for i in range(n)]
    res = []
    for i in div:
        res.append(max(f(np.linspace(i, i + dx, 1000))) * dx)
    return sum(res)

print('S(M_i * dx_i) =', darb_hi(lambda x: x, 1, 2, 4))
```

```python
# upr 5

def f(x):
    return np.exp((-1) * x**2)

print('ис при значениях левой границы:', isum_left(f, 1, 2, 1000))
print('ис при значениях правой границы:', isum_right(f, 1, 2, 1000))
print('нижняя сумма:', darb_lo(f, 1, 2, 1000))
print('верхняя сумма:', darb_hi(f, 1, 2, 1000))
```

```python
darb_hi(f, 1, 2, 100000) - darb_lo(f, 1, 2, 100000)
```

```python
# upr 6

x1 = np.linspace(1, 2, 100)
x2 = np.linspace(1, 2, 1000)

simpson1 = scipy.integrate.simpson(np.exp(-x1**2), x1)
simpson2 = scipy.integrate.simpson(np.exp(-x2**2), x2)

dl1 = darb_lo(lambda x: np.exp(-x**2), 1, 2, 100)
dl2 = darb_lo(lambda x: np.exp(-x**2), 1, 2, 1000)
dh1 = darb_hi(lambda x: np.exp(-x**2), 1, 2, 100)
dh2 = darb_hi(lambda x: np.exp(-x**2), 1, 2, 1000)

ls1 = isum_left(lambda x: np.exp(-x**2), 1, 2, 100)
ls2 = isum_left(lambda x: np.exp(-x**2), 1, 2, 1000)
rs1 = isum_right(lambda x: np.exp(-x**2), 1, 2, 100)
rs2 = isum_right(lambda x: np.exp(-x**2), 1, 2, 1000)
```

```python
print('формула Симпсона:')
print('\t100:', simpson1, '\n\t1000:', simpson2)
print('суммы Дарбу:')
print('\tнижние:\n', '\t\t100:', dl1, '\n\t\t1000:', dl2)
print('\tверхние:\n', '\t\t100:', dh1, '\n\t\t1000:', dh2)
print('равномерное разбиение')
print('\tлевая граница:')
print('\t\t100:', ls1, '\n\t\t1000:', ls2)
print('\tправая граница:')
print('\t\t100:', rs1, '\n\t\t1000:', rs2)
```

```python
print('погрешности (в порядке, как выше), далее для 1000 разбиений')
print(np.abs(simpson1 - simpson1))
print(np.abs(simpson2 - simpson1))
print(np.abs(dl1 - simpson1))
print(np.abs(dl2 - simpson1))
print(np.abs(dh1 - simpson1))
print(np.abs(dh2 - simpson1))
print(np.abs(ls1 - simpson1))
print(np.abs(ls2 - simpson1))
print(np.abs(rs1 - simpson1))
print(np.abs(rs2 - simpson1))
```

```python
# C1

sp.integrate(1 / sp.sqrt(x**2 + 2 * x + 3))
```

вручную:  

$\int \frac{dx}{\sqrt{x^2 + 2x + 3}} = \int{\frac{dx}{\sqrt{(x + 1)^2 + 2}}} = 
\int{\frac{d(x + 1)}{\sqrt{(x + 1)^2 + 2}}} = \ln |x + 1 + \sqrt{(x + 1)^2 + 2}| + C$  
$\ln |x + 1 + \sqrt{(x + 1)^2 + 2}| = \ln |x + 1 + \sqrt{2} \sqrt{\big(\frac{x + 1}{\sqrt{2}}\big)^2 + 1}| = \ln |\frac{x + 1}{\sqrt{2}} \sqrt{2} + \sqrt{2} \sqrt{\big(\frac{x + 1}{\sqrt{2}}\big)^2 + 1}| = \ln{\sqrt{2}} + \ln |\frac{x + 1}{\sqrt{2}} + \sqrt{\big(\frac{x + 1}{\sqrt{2}}\big)^2 + 1}| = arsh(\frac{x + 1}{\sqrt{2}}) + C$

```python
# C3

def isum_l(f, a, b, n, l):
    dx = (b - a) / n
    div = [a + i * dx for i in range(n + 1)]
    points = [f((b_i * l + a_i) / (l + 1)) for a_i, b_i in zip(div[:-1], div[1:])]
    return sum([i * dx for i in points])

isum_l(lambda x: x, 1, 2, 4, 1)
```
