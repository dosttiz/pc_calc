```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math
```

$$ \int \int (x + y^2) dx dy = \int_0^2 dx \int_{x^2}^{2x} (x + y^2)dy = 
\int_0^4 dy \int_{y / 2}^{\sqrt y} (x + y^2)dx  $$

```python
# upr 1

x = sp.Symbol('x')
y = sp.Symbol('y')
x1 = np.linspace(-3, 3)
plt.plot(x1, 2 * x1)
plt.plot(x1, x1**2)
plt.grid()
i1 = sp.integrate(sp.integrate(x + y**2, (y, x**2, 2 * x)), (x, 0, 2))
i2 = sp.integrate(sp.integrate(x + y**2, (x, y / 2, sp.sqrt(y))), (y, 0, 4))
print('Первый способ', i1)
print('Второй способ', i2)
```

$$ \int \int \int z dx dy dz = \int_0^1 dx \int_0^{1 - x} dy \int_0^{1 - x - y}zdz $$

```python
# upr 2

from matplotlib import cm
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 1 - X - Y
ax.plot_surface(X, Y, Z, cmap=cm.cool, linewidth=0)
xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = 0 * X
ax.plot_wireframe(X, Y, Z, rstride = 2, cstride = 2)
xs = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
X, Z = np.meshgrid(xs, zs)
Y = 0 # X - X + Z - Z
ax.plot_wireframe(X, Y, Z, rstride = 2, cstride = 2)
ys = np.linspace(0, 1, 50)
zs = np.linspace(0, 1, 50)
Y, Z = np.meshgrid(ys, zs)
X = 0 # Y - Y + Z - Z
ax.plot_wireframe(X, Y, Z, rstride = 2, cstride = 2)

x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')
i = sp.integrate(sp.integrate(sp.integrate(z, (z, 0, 1 - x - y)), (y, 0, 1 - x)), (x, 0, 1))
print(i)
```

$$ \int \int x \sqrt{1 - y} dx dy = \int_0^1 dx \int_0^{x^2} x \sqrt{1 - y} dy =
\int_0^1 dy \int_{\sqrt y}^1 x \sqrt{1 - y} dx $$

```python
# upr 3

x1 = np.linspace(0, 1)
plt.plot(x1, x1**2)
plt.grid()
plt.axvline(1)
plt.axhline(0)

x = sp.Symbol('x')
y = sp.Symbol('y')
i1 = sp.integrate(sp.integrate(x * sp.sqrt(1 - y), (y, 0, x**2)), (x, 0, 1))
i2 = sp.integrate(sp.integrate(x * sp.sqrt(1 - y), (x, sp.sqrt(y), 1)), (y, 0, 1))
print('Первый способ', i1)
print('Второй способ', i2)
```
