```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# upr 1

def part_sum(a, n):
    s_i = [a(i) for i in range(1, n + 1)]
    s_i = [sum(s_i[:i]) for i in range(len(s_i))]
    plt.plot(range(n), s_i, 'o--')
    # return s_i
```

```python
part_sum(lambda n: n, 1000)
part_sum(lambda n: 1 / n, 1000)
plt.grid()
```
